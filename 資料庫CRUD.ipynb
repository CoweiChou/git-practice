```python
import pymysql
from prettytable import PrettyTable
import IPython.display as dp


c1 = pymysql.connect(
    host=input("請輸入主機位置："),
    user=input("請輸入資料庫帳號："),
    passwd=input("請輸入資料庫密碼："),
    charset="utf8",
    port=int(input("請輸入資料庫的port：")),
    db=input("請輸入資料庫名稱："),
)
p1 = c1.cursor()

dp.clear_output(True)

table_name = input("請輸入資料表名稱：")
table = PrettyTable(["id", "name", "species", "type", "gender", "size", "cat_color", "age", "sterilization", "address", "url", "image_url", "location", "details"], encoding="utf-8")


def new_table():
        p1.execute(f"SELECT * FROM `{table_name}`")
        results = p1.fetchall()
        table.clear_rows()
        for row in results:
            table.add_row(row)
        print(table)


def n_action():
    print("(0) 離開程式\n(1) 顯示列表\n(2) 新增資料\n(3) 修改資料\n(4) 刪除資料")
    return int(input("指令："))
    dp.clear_output(True)


while True:
    try:
        action = n_action()
        dp.clear_output(True)

        if action == 0:
            break

        elif action == 1:
            new_table()



        elif action == 2:
            p1.execute(f"ALTER TABLE `{table_name}` MODIFY id INT AUTO_INCREMENT;")
            c1.commit()

            p1.execute(
                f"INSERT INTO `{table_name}` (`name`,`species`,`type`,`gender`,`size`,`cat_color`,`age`,`sterilization`,`address`,`url`,`image_url`,`location`,`details`) VALUES(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)",
                [input("請輸入貓咪名稱："),
                 input("請輸入貓咪類型："),
                 input("請輸入貓："),
                 input("請輸入公貓或母貓："),
                 input("請輸入小型/中型/大型："),
                 input("請輸入貓咪花色："),
                 input("請輸入成貓或幼貓："),
                 input("請輸入已結紮或未結紮："),
                 input("請輸入縣市 鄉鎮區："),
                 input("請輸入url："),
                 input("請輸入image_url："),
                 input("請輸入北部/中部/南部："),
                 input("請輸入貓咪個性描述："),
                 ]
            )
            dp.clear_output(True)
            c1.commit()



        elif action == 3:
            new_table()
            n = input("請選擇你要修改的資料編號：")


            p1.execute(f"SELECT * FROM `{table_name}` WHERE `id` = %s", [n])
            result = p1.fetchone()

            if not result:
                print("找不到編號為", n, "的資料")
                dp.clear_output(wait=True)
                continue


            result_dict = {
                "name": result[1],
                "species": result[2],
                "type": result[3],
                "gender": result[4],
                "size": result[5],
                "cat_color": result[6],
                "age": result[7],
                "sterilization": result[8],
                "address": result[9],
                "url": result[10],
                "image_url": result[11],
                "location": result[12],
                "details": result[13]
            }



            print("原始資料：")
            print("1. 貓咪名稱:", result_dict["name"])
            print("2. 貓咪類型:", result_dict["species"])
            print("3. 貓:", result_dict["type"])
            print("4. 貓咪性別:", result_dict["gender"])
            print("5. 貓咪體型:", result_dict["size"])
            print("6. 貓咪花色:", result_dict["cat_color"])
            print("7. 貓咪年紀:", result_dict["age"])
            print("8. 結紮狀況:", result_dict["sterilization"])
            print("9. 縣市:", result_dict["address"])
            print("10. url:", result_dict["url"])
            print("11. image_url:", result_dict["image_url"])
            print("12. 地區:", result_dict["location"])
            print("13. 個性描述:", result_dict["details"])

            # 讓用戶輸入要修改的欄位和值，如果沒有輸入則保留原始值
            columns_to_update = ["name", "species", "type", "gender", "size", "cat_color", "age", "sterilization", "address", "url", "image_url", "location", "details"]
            column_index = input("請輸入要修改的欄位編號（留空表示不修改）：")
            if column_index:
                column = columns_to_update[int(column_index) - 1]
                user_input = input("請輸入要修改的內容：")
                if user_input:
                    result_dict[column] = user_input


            p1.execute(
                f"UPDATE `{table_name}` SET `name`= %s,`species`= %s,`type`= %s, `gender`= %s,`size`= %s,`cat_color`= %s, `age`= %s,`sterilization`= %s,`address`= %s, `url`= %s,`image_url`= %s,`location`= %s, `details`= %s WHERE `id` = %s",
                [result_dict["name"], result_dict["species"], result_dict["type"], result_dict["gender"], result_dict["size"], result_dict["cat_color"], result_dict["age"], result_dict["sterilization"], result_dict["address"], result_dict["url"], result_dict["image_url"], result_dict["location"], result_dict["details"], n]
            )
            c1.commit()
            dp.clear_output(True)




        elif action == 4:
            new_table()
            n = input("請選擇你要刪除的資料編號：")
            p1.execute(
                f"DELETE FROM `{table_name}` WHERE `id` = %s", [n])
            c1.commit()
            dp.clear_output(True)


    except KeyboardInterrupt:
        break

c1.close()



```
